###Chargement de l'extension Octave magic

```python
%load_ext oct2py.ipython
```

###<span style="color:red">Définition et tracé d'une fonction 3D complexe dans [Octave](http://www.gnu.org/software/octave/)</span> <img src="http://enacit1.epfl.ch/cours_matlab/octave_logo.gif">

###$$\sin(x - 0.5)^2 + cos(y - 0.5)^2$$



```python
%%octave -s 1400,467 -f png

subplot(121);
[x, y] = meshgrid(0:0.1:3);
r = sin(x - 0.5).^2 + cos(y - 0.5).^2;
surf(x, y, r);
```

###<span style="color:red"> Transfert des variables Octave à des variables Python</span>

```python
%octave_pull x y r
```

###Chargement de l'extension R magic

```python
load_ext rpy2.ipython
```

###<span style="color:red">Transfert des variables Python à des variables R</span>
<img src="http://www.r-project.org/Rlogo.jpg">

```python
%Rpush x y r 
```

###Tracé de la fonction avec le package [plot3D](http://cran.r-project.org/web/packages/plot3D/index.html)

```python
%%R
library(plot3D)
surf3D(x = x,
       y = y,
       z = r ,
       colkey=FALSE,
       bty="b2",
       main="fonction définie dans Octave et tracée avec R")
```

###<span style="color:red">Transfert des variables R à des variables Python</span>

<img src="https://www.python.org/static/img/python-logo.png">

```python
%Rpull x y r 
```

```python
%matplotlib inline
```

```python
from mpl_toolkits.mplot3d.axes3d import *
import matplotlib.pyplot as plt
from matplotlib import cm
```

###Tracé de la fonction avec le module [matplotlib](http://matplotlib.org/)

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x, y, r, rstride=1, cstride=1, cmap=cm.jet,linewidth=1, antialiased=True)
plt.show()
```
